```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [100, 250, 500, 750, 1000, 2000, 3000, 4000, 5000, 6000]
res = [148.3, 421.9, 654.35, 292, 993.60, 1127.93, 1615.35, 1104, 1358.62, 1300.91]
plt.plot(x, res,marker='o')
yerr=[105.4, 308, 289.34, 292, 323.5, 433.26, 757, 576.7, 467.5, 295.8]
print(len(yerr), len(x), len(res))
plt.errorbar(x, res, yerr)
plt.ylabel('Policy Average Return')
plt.xlabel('Num Agent Train Steps per Iter')
plt.title("Num training steps vs Policy Return")
plt.savefig(dpi='figure', fname='Hopper training steps sweep.png', format='png')
plt.show()
```

```python
iter = np.arange(10)
ant = [3973, 4175, 4796, 4636, 4680, 4698, 4740, 4337, 4635, 4677]
hopper = [993.6, 2811, 2856, 3713, 3717, 3099, 3718, 3715, 3711, 3713]
ant_err=[815, 1067, 65.5, 69.7, 91, 79.5, 103, 955, 111.7, 60.4]
hopper_err=[323.5, 632.2, 609.5, 6.57, 3.55, 1023, 3.3, 2.23, 4.177, 1.58]

name = 'Hopper'
data = hopper
err = hopper_err
exp_perf = 3717.5
bc_perf = 993.6

print(len(err), len(iter), len(data))
plt.plot(iter, data,marker='o', label='Dagger')
plt.plot(iter, [exp_perf] * len(iter), label='Expert return')
plt.plot(iter, [bc_perf] * len(iter), label='BC return')

plt.errorbar(iter, data, err)
plt.legend()
plt.ylabel('Policy Average Return')
plt.xlabel('Num Dagger Iter')
plt.title(name+": Num Dagger iter vs Policy Return")
plt.savefig(dpi='figure', fname=name+'_dagger.png', format='png')
plt.show()
```

```python
name = 'Ant'
data = ant
err = ant_err
exp_perf = 4681.9
bc_perf = 3973.07

print(len(err), len(iter), len(data))
plt.plot(iter, data,marker='o', label='Dagger')
plt.plot(iter, [exp_perf] * len(iter), label='Expert return')
plt.plot(iter, [bc_perf] * len(iter), label='BC return')

plt.errorbar(iter, data, err)
plt.legend()
plt.ylabel('Policy Average Return')
plt.xlabel('Num Dagger Iter')
plt.title(name+": Num Dagger iter vs Policy Return")
plt.savefig(dpi='figure', fname=name+'_dagger.png', format='png')
plt.show()
```
